```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Titanic-Dataset.csv")

print("First 5 rows:\n", df.head())
print("\nInfo:\n")
print(df.info())
print("\nSummary statistics:\n", df.describe())
print("\nCategorical summary:\n", df.describe(include='object'))
print("\nMissing values:\n", df.isnull().sum())
```

```python
df[['Age', 'Fare']].hist(bins=20, figsize=(10,5))
plt.suptitle("Histograms for Age and Fare")
plt.show()

# Boxplot: Age distribution by Survival
plt.figure(figsize=(8,5))
sns.boxplot(x='Survived', y='Age', data=df)
plt.title('Age distribution by Survival')
plt.show()
```

```python
# Boxplot: Fare distribution by Pclass
plt.figure(figsize=(8,5))
sns.boxplot(x='Pclass', y='Fare', data=df)
plt.title('Fare distribution by Passenger Class')
plt.show()
```

```python
# Scatter plot: Age vs Fare colored by Survival
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='Age', y='Fare', hue='Survived', palette='Set1')
plt.title('Age vs Fare by Survival')
plt.show()
```

```python
# Pairplot for selected features
sns.pairplot(df[['Age', 'Fare', 'Survived', 'Pclass']], hue='Survived', palette='Set2')
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Feature Correlation Heatmap')
plt.show()
```

```python
print("\nBasic Inferences:")
print("- Younger passengers and females tend to have higher survival rate.")
print("- Higher ticket fares (likely first-class) correlate with higher survival.")
print("- Family size might affect survival (alone vs with family).")
```
