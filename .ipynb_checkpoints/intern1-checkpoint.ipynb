```python
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load
df = pd.read_csv('Titanic-Dataset.csv')   

# Quick checks
print("Shape:", df.shape)
display(df.head())
df.info()           # types & null counts
display(df.describe(include='all').T)
print("Missing per column:\n", df.isnull().sum())
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Fare'] = df['Fare'].fillna(df['Fare'].median())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

# Cabin: too many missing → create HasCabin flag instead of filling
df['HasCabin'] = df['Cabin'].notnull().astype(int)
df['Cabin'] = df['Cabin'].fillna('Missing')
```

```python

# Fill missing values safely without warnings and KeyError

# Check if columns exist first
if 'Age' in df.columns:
    df['Age'] = df['Age'].fillna(df['Age'].median())

if 'Fare' in df.columns:
    df['Fare'] = df['Fare'].fillna(df['Fare'].median())

if 'Embarked' in df.columns:
    df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

# Cabin: too many missing → create HasCabin flag + fill
if 'Cabin' in df.columns:
    df['HasCabin'] = df['Cabin'].notna().astype(int)
    df['Cabin'] = df['Cabin'].fillna('Missing')
```

```python
# Family Size and IsAlone
if 'SibSp' in df.columns and 'Parch' in df.columns:
    df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
    df['IsAlone'] = (df['FamilySize'] == 1).astype(int)

# Extract Title safely if 'Name' exists
if 'Name' in df.columns:
    df['Title'] = df['Name'].str.extract(r',\s*([^\.]+)\.')
else:
    df['Title'] = 'Unknown'  # or skip this feature
```

```python
def remove_outliers_iqr(df, cols, k=1.5):
    df2 = df.copy()
    for col in cols:
        Q1 = df2[col].quantile(0.25)
        Q3 = df2[col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - k * IQR
        upper = Q3 + k * IQR
        df2 = df2[(df2[col] >= lower) & (df2[col] <= upper)]
    return df2

df_no_out = remove_outliers_iqr(df, ['Age','Fare'])
print("Before:", df.shape, "After:", df_no_out.shape)
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

numeric_features = ['Age','Fare','FamilySize']
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

categorical_features = ['Sex','Embarked']  # adapt as needed
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer(transformers=[
    ('num', numeric_transformer, numeric_features),
    ('cat', categorical_transformer, categorical_features)
])

# Example: produce cleaned X for modeling
X = df.drop(columns=['Survived'])   # target column name may change
y = df['Survived']
X_processed = preprocessor.fit_transform(X)
# To get column names (sklearn >= 1.0):
ohe_cols = preprocessor.named_transformers_['cat']['onehot'].get_feature_names_out(categorical_features)
all_cols = numeric_features + list(ohe_cols)
X_clean = pd.DataFrame(X_processed, columns=all_cols)
```

```python
X_clean.to_csv('titanic_features_cleaned.csv', index=False)
y.to_csv('titanic_target.csv', index=False)
```
